```python
from bokeh.plotting import show, figure, output_notebook
from math import exp
from random import uniform
output_notebook()

def gen_sigmoid(x):
    y = []
    w0 = uniform(-1,1)
    w1 = uniform(-1,1)

    for i in x:
        v = 1/(1.0 + exp(-1.0 * (w0 + w1*i)))
        y.append(v)
    return y


x = [i for i in range(-10, 10)]

# Layer-1
s1 = gen_sigmoid(x)
s2 = gen_sigmoid(x)
s3 = gen_sigmoid(x)
s4 = gen_sigmoid(x)


f1 = figure(title='sigmoids', height=600, width=700)

f1.line(x, s1, color='green', line_width=3, legend_label='s1')
f1.line(x, s2, color='red', line_width=3, legend_label='s2')
f1.line(x, s3, color='yellow', line_width=3, legend_label='s3')
f1.line(x, s4, color='black', line_width=3, legend_label='s4')



show(f1)
```

```python

# Layer-2: add s1, s2, s3, s4
s1_4 = [sum(x) for x in zip(s1, s2, s3, s4)]
s5 = gen_sigmoid(s1_4)
s6 = gen_sigmoid(s1_4)
s7 = gen_sigmoid(s1_4)
s8 = gen_sigmoid(s1_4)

f2 = figure(title='sigmoids', height=600, width=700)

f2.line(x, s5, color='green', line_width=3, legend_label='s5')
f2.line(x, s6, color='red', line_width=3, legend_label='s6')
f2.line(x, s7, color='yellow', line_width=3, legend_label='s7')
f2.line(x, s8, color='black', line_width=3, legend_label='s8')

show(f2)
```

```python
# Layer-3: add s5, s26, s7, s8
s5_8 = [sum(x) for x in zip(s5, s6, s7, s8)]
s9 = gen_sigmoid(s5_8)
s10 = gen_sigmoid(s5_8)
s11 = gen_sigmoid(s5_8)
s12 = gen_sigmoid(s5_8)

f3 = figure(title='sigmoids', height=600, width=700)

f3.line(x, s9, color='green', line_width=3, legend_label='s9')
f3.line(x, s10, color='red', line_width=3, legend_label='s10')
f3.line(x, s11, color='yellow', line_width=3, legend_label='s11')
f3.line(x, s12, color='black', line_width=3, legend_label='s12')

show(f3)
```

```python
# Final Output: add s9, s10, s11, s12
s9_12 = [sum(x) for x in zip(s9, s10, s11, s12)]
y = gen_sigmoid(s9_12)

f4 = figure(title='sigmoids', height=600, width=700)
f4.line(x, y, color='red', line_width=3, legend_label='Y')

show(f4)
```
