For each person in the health metrics dataset, find the other most similar person. Use proper similarity metrics for the given attribute types.



```python
import numpy as np
from scipy.spatial.distance import cosine

data = [
    [1, 0, 0, 24, 60, 120],
    [1, 1, 1, 28, 62, 125],
    [0, 0, 0, 20, 55, 110],
    [0, 1, 1, 30, 70, 130],
    [1, 0, 1, 31, 69, 115]
]

for i in range(len(data)):
    p1 = data[i]
    p1Binary = p1[:3]
    p1Cont = p1[3:]

    max_similarity = -np.inf
    max_j = -1

    for j in range(len(data)):
        if i != j:
            
            p2 = data[j]
            p2Binary = p2[:3]
            p2Cont = p2[3:]

            smc = np.sum(np.array(p1Binary) == np.array(p2Binary)) / len(p1Binary)

            numerator = np.dot(p1Cont, p2Cont)
            denominator = np.sqrt(np.sum(np.square(p1Cont))) * np.sqrt(np.sum(np.square(p2Cont)))
            cos_sim = numerator / denominator

            similarity = (smc + cos_sim) / 2

            if similarity > max_similarity:
                max_similarity = similarity
                max_j = j

    print(f"Person {i+1} is most similar to person {max_j+1} with a similarity score of {max_similarity}")
```
